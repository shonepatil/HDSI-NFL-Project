```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

low_memory = False

# Read csv into pandas dataframe
df = pd.read_csv('train.csv')
```

```python
df
```

```python
pd.set_option('display.max_columns', None)
positions = df.Position.unique()
defensive_positions = positions[np.array([0, 1, 2, 3, 4, 5, 13, 14, 16, 17, 22, 23, 24])]
df_defense = df[df['Position'].isin(defensive_positions)]#filter by defense
df_defense
```

```python
avg_speed_defense = df_defense.groupby('PlayId').mean()[['S', 'Yards']]

plt.figure(figsize=(20,10))
plt.title('Average Defender Speed at Handoff Vs. Yards Gained')
plt.xlabel('Average Defender Speed at Handoff')
plt.ylabel('Yards Gained')
plt.scatter(avg_speed_defense['S'], avg_speed_defense['Yards'])
```

```python
avg_accel_defense = df_defense.groupby('PlayId').mean()[['A', 'Yards']]

plt.figure(figsize=(20,10))
plt.title('Average Defender Acceleration at Handoff Vs. Yards Gained')
plt.xlabel('Average Defender Acceleration at Handoff')
plt.ylabel('Yards Gained')
plt.scatter(avg_accel_defense['A'], avg_accel_defense['Yards'])
```

```python
defenders_in_box = df_defense.groupby('PlayId').mean()[['DefendersInTheBox', 'Yards']]

plt.figure(figsize=(10,5))
plt.title('Number of Defenders in Box Vs. Yards Gained')
plt.xlabel('Number of Defenders in Box')
plt.ylabel('Yards Gained')
plt.scatter(defenders_in_box['DefendersInTheBox'], defenders_in_box['Yards'])
```

```python
avg_dis_moved_defense = df_defense.groupby('PlayId').mean()[['Dis', 'Yards']]

plt.figure(figsize=(20,10))
plt.title('Average Defender Distance Moved Since Snap Vs. Yards Gained')
plt.xlabel('Average Defender Distance Moved Since Snap')
plt.ylabel('Yards Gained')
plt.scatter(avg_dis_moved_defense['Dis'], avg_dis_moved_defense['Yards'])
```
