```python
miss_pos_19_mod = []
miss_pos_18_mod = []
miss_pos_17_mod = []

for i in nfl_main_18['position'].drop_duplicates().to_list():
    if i not in nfl_18_mod['position'].to_list():
        miss_pos_18_mod.append(i)
        
for j in nfl_main_17['position'].drop_duplicates().to_list():
    if j not in nfl_17_mod['position'].to_list():
        miss_pos_17_mod.append(j)

for j in nfl_main_19['position'].drop_duplicates().to_list():
    if j not in nfl_19_mod['position'].to_list():
        miss_pos_19_mod.append(j)

print("Number of Mismatched Positions in 19: " + str(len(miss_pos_19_mod)))
print("Number of Mismatched Positions in 18: " + str(len(miss_pos_18_mod)))
print("Number of Mismatched Positions in 17: " + str(len(miss_pos_17_mod)))
```

```python
# Function used to generate missing values for players
# Player positions not affected by this function
def position_reassignment(i):
    if i == "HB" or i == "FB":
        return "RB"
    if i == "OT" or i == "T" or i == "RT" or i == "LT":
        return "T"
    if i == "G" or i == "OG" or i == "RG" or i == "LG":
        return "G"
    if i == "ILB" or i == "OLB" or i == "LB" or i == "LOLB" or i == "ROLB" or i == "DL" or i == "MLB":
        return "LB"
    if i == "CB" or i == "DB":
        return "DB"
    if i == "LE" or i == "RE" or i == "DE":
        return "DE"
    if i == "NT" or i == "DT":
        return "DT"
    if i == "SS" or i == "S" or i == "SAF":
        return "S"
    else:
        return i 
```
