```python
def score(submission):
    df = pd.read_csv('train.csv')
    submission = submission.set_index('PlayId')
    def score_row(row, act_yards):
        score = []
        for i in range(-99, 100):
            if i >= act_yards:
                score.append((row[i + 99] - 1) ** 2)
            else:
                score.append((row[i + 99]) ** 2)
        play_score = sum(score) / 199
        return play_score
    play_scores_sum = 0
    for i in range(len(submission)):
        play_scores_sum += score_row(submission.iloc[i], df[df['PlayId'] == submission.index[i]].iloc[0]['Yards'])
    final_score = play_scores_sum / len(submission)
    return final_score
        
```

```python
score(yard_hist_sub)
```
