```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

low_memory = False

# Read csv into pandas dataframe
df = pd.read_csv('train.csv')
```

```python
import math

target_play_id = 20181230154157
curr_play = df[df['PlayId'] == target_play_id]

plt.figure(figsize=(20,10))
plt.ylim(top = 53.3, bottom = 0)
plt.scatter(curr_play['X'][:11], curr_play['Y'][:11], s = 100, color = 'Red')#plot defense
plt.scatter(curr_play['X'][11:], curr_play['Y'][11:], s = 100, color = 'Blue')#plot offense
plt.axvline(x = 120 - (curr_play['YardLine'].iloc[0] + 10), label = "Line of Scrimmage", ls = "--")#plot line of scrimmage
plt.legend()
plt.text(120 - (curr_play['YardLine'].iloc[0] + 10) + plt.xlim()[1] * 0.01 , plt.ylim()[1] * 0.01,\
        str(curr_play['YardLine'].iloc[0]),\
        horizontalalignment='center', fontweight='bold', color='black')

#plot directional vectors at start of play
for i in range(len(curr_play)):
    plt.quiver(curr_play['X'].iloc[i], curr_play['Y'].iloc[i], math.cos(math.radians(-(curr_play['Dir'].iloc[i] - 90))), \
           math.sin(math.radians(-(curr_play['Dir'].iloc[i] - 90))), width = 0.003)
```
